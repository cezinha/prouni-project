```python
import pandas as pd
```

## Reading data from CSV

```python
df = pd.read_csv("cursos-prouni.csv")
```

```python
df.head()
```

### Get columns labels

```python
for i, v in enumerate(df.columns):
    print(i, v)
```

### Get courses listing

```python
df.nome.unique()
```

# Course: Administração / Administração de Empresas 
(Business Administration)

```python
df[(df.nome == "Administração") | (df.nome == "Administração de Empresas")]['mensalidade'].describe()
```

### Monthly payment value by UF (state)

```python
df[(df.nome == "Administração") | (df.nome == "Administração de Empresas")].groupby("uf_busca")['mensalidade'].describe()
```

### Get shifts from 'Administração' (Business Administration) course
To check if has 'Curso a Distância' (Distance Education)

```python
df[(df.nome == "Administração") | (df.nome == "Administração de Empresas")].turno.unique()
```

###  Monthly payment value by UF (state) without 'Curso a Distância' (Distance Education)

```python
df[((df.nome == "Administração") | (df.nome == "Administração de Empresas")) & (df.turno != "Curso a Distância")].groupby("uf_busca")['mensalidade'].describe()
```

### Monthly payment value: Histogram without 'Curso a distância' (Distance Education)


```python
%matplotlib notebook

df[((df.nome == "Administração") | (df.nome == "Administração de Empresas")) & (df.turno != "Curso a Distância")]['mensalidade'].hist()
```

### Monthiy payment value: Boxplot Brazil - without 'Curso a Distância' (Distance Education)

```python
%matplotlib notebook

matplotlib.style.use('ggplot')
df[((df.nome == "Administração") | (df.nome == "Administração de Empresas")) & (df.turno != "Curso a Distância")].boxplot(column="mensalidade")
```

### Monthly payment value: Boxplot by UF - without 'Curso a Distância' (Distance Education)

```python
%matplotlib notebook

matplotlib.style.use('ggplot')
df[((df.nome == "Administração") | (df.nome == "Administração de Empresas")) & (df.turno != "Curso a Distância")].boxplot(column="mensalidade", by="uf_busca")
```
