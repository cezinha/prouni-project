```python
import pandas as pd
```

## Lendo os dados do CSV

```python
df = pd.read_csv("cursos-prouni.csv")
```

```python
df.head()
```

### Lista os campos

```python
for i, v in enumerate(df.columns):
    print(i, v)
```

# Curso: Administração

```python
df[df.nome == "Administração"]['mensalidade'].describe()
```

### Mensalidade por UF

```python
df[df.nome == "Administração"].groupby("uf_busca")['mensalidade'].describe()
```

### Lista os turnos do curso de Administração
Para conferir se tem o curso a Distância

```python
df[(df.nome == "Administração")].turno.unique()
```

### Mensalidade por UF sem o Curso a Distância

```python
df[(df.nome == "Administração") & (df.turno != "Curso a Distância")].groupby("uf_busca")['mensalidade'].describe()
```

### Mensalidade: Histogram (sem curso a distância)

```python
%matplotlib notebook

df[(df.nome == "Administração") & (df.turno != "Curso a Distância")]['mensalidade'].hist()
```

### Mensalidade: Boxplot (Brasil - sem curso a distância)

```python
%matplotlib notebook

matplotlib.style.use('ggplot')
df[(df.nome == "Administração") & (df.turno != "Curso a Distância")].boxplot(column="mensalidade")
```

```python
%matplotlib notebook

matplotlib.style.use('ggplot')
df[(df.nome == "Administração") & (df.turno != "Curso a Distância")].boxplot(column="mensalidade", by="uf_busca")
```

# Curso: Ciência da Computação ou Ciências da Computação

```python
df[(df.nome == "Ciência da Computação") | (df.nome == "Ciências da Computação")]['mensalidade'].describe()
```

### Mensalidade por UF

```python
df[(df.nome == "Ciência da Computação") | (df.nome == "Ciências da Computação")].groupby("uf_busca")['mensalidade'].describe()
```

### Mensalidade: Histogram

```python
%matplotlib notebook

df[(df.nome == "Ciência da Computação") | (df.nome == "Ciências da Computação")]['mensalidade'].hist()
```

### Mensalidade: Boxplot (Brasil)

```python
%matplotlib notebook

matplotlib.style.use('ggplot')
df[(df.nome == "Ciência da Computação") | (df.nome == "Ciências da Computação")].boxplot(column="mensalidade")
```

### Mensalidade: Boxplot (por UF)

```python
%matplotlib notebook

matplotlib.style.use('ggplot')
df[(df.nome == "Ciência da Computação") | (df.nome == "Ciências da Computação")].boxplot(column="mensalidade", by="uf_busca")
```

### Lista os turnos do curso de Ciência da Computação
Para conferir se existe o Curso a Distância

```python
df[((df.nome == "Ciência da Computação") | (df.nome == "Ciências da Computação"))].turno.unique()
```
